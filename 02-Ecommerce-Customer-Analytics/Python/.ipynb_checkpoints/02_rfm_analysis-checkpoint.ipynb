```python

import pandas as pd
import numpy as np
from datetime import timedelta
import matplotlib.pyplot as plt
import seaborn as sns


master_df = pd.read_csv('/Users/poornavenkat/Documents/GitHub/SQL-Projects/02-Ecommerce-Customer-Analytics/data/Output/master_dataset.csv')

master_df['order_purchase_timestamp'] = pd.to_datetime(master_df['order_purchase_timestamp'])


df_delivered = master_df[master_df['order_status'] == 'delivered'].copy()
print(f"Working with {len(df_delivered)} delivered orders")
```

```python

snapshot_date = df_delivered['order_purchase_timestamp'].max() + timedelta(days=1)
print(f"Snapshot date for analysis: {snapshot_date}")

# Calculate RFM
rfm = df_delivered.groupby('customer_id').agg({
    'order_purchase_timestamp': lambda x: (snapshot_date - x.max()).days,  # Recency
    'order_id': 'count',  # Frequency  
    'payment_total': 'sum'  # Monetary
})

rfm.columns = ['Recency', 'Frequency', 'Monetary']
rfm = rfm[rfm['Monetary'] > 0]  # Remove zero/negative monetary values

print(f"RFM table created for {len(rfm)} customers")
print("\nRFM Statistics:")
print(rfm.describe())
```

```python

rfm['R_Quartile'] = pd.qcut(rfm['Recency'], 4, labels=['4','3','2','1'])  # Lower is better for recency
rfm['F_Quartile'] = pd.qcut(rfm['Frequency'].rank(method='first'), 4, labels=['1','2','3','4'])
rfm['M_Quartile'] = pd.qcut(rfm['Monetary'], 4, labels=['1','2','3','4'])

rfm['RFM_Score'] = rfm['R_Quartile'].astype(str) + rfm['F_Quartile'].astype(str) + rfm['M_Quartile'].astype(str)

print("RFM Scores calculated!")
print("\nTop 10 customers by RFM Score:")
print(rfm.nlargest(10, 'Monetary')[['Recency', 'Frequency', 'Monetary', 'RFM_Score']])
```

```python
def segment_customers(row):
    if row['RFM_Score'] in ['444', '443', '434', '344']:
        return 'Champions'
    elif row['RFM_Score'] in ['442', '441', '433', '432', '343', '342', '333', '334']:
        return 'Loyal Customers'
    elif row['RFM_Score'] in ['431', '421', '331', '321', '422', '332']:
        return 'Potential Loyalists'
    elif row['RFM_Score'] in ['411', '412', '311', '312', '313']:
        return 'New Customers'
    elif row['RFM_Score'] in ['424', '423', '413', '414', '324', '323', '314']:
        return 'Promising'
    elif row['RFM_Score'] in ['341', '241', '242', '141', '142', '143']:
        return 'Need Attention'
    elif row['RFM_Score'] in ['244', '234', '233', '134', '133', '144']:
        return 'About to Sleep'
    elif row['RFM_Score'] in ['222', '223', '122', '123', '132', '231', '232', '221']:
        return 'At Risk'
    elif row['RFM_Score'] in ['211', '212', '121', '131', '112', '111']:
        return 'Cannot Lose Them'
    else:
        return 'Lost'

rfm['Segment'] = rfm.apply(segment_customers, axis=1)

print("Customer Segmentation Complete!")
print("\nSegment Distribution:")
print(rfm['Segment'].value_counts())
```

```python

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Segment distribution
rfm['Segment'].value_counts().plot(kind='bar', ax=axes[0,0], color='steelblue')
axes[0,0].set_title('Customer Distribution by Segment')
axes[0,0].set_xlabel('Segment')
axes[0,0].set_ylabel('Number of Customers')

# Recency vs Frequency
axes[0,1].scatter(rfm['Recency'], rfm['Frequency'], alpha=0.5, c=rfm['Monetary'], cmap='viridis')
axes[0,1].set_title('Recency vs Frequency (colored by Monetary)')
axes[0,1].set_xlabel('Recency (days)')
axes[0,1].set_ylabel('Frequency (orders)')

# Average monetary by segment
segment_monetary = rfm.groupby('Segment')['Monetary'].mean().sort_values(ascending=False)
segment_monetary.plot(kind='barh', ax=axes[1,0], color='green')
axes[1,0].set_title('Average Monetary Value by Segment')
axes[1,0].set_xlabel('Average Monetary Value ($)')

# Recency distribution
axes[1,1].hist(rfm['Recency'], bins=50, color='coral', edgecolor='black')
axes[1,1].set_title('Recency Distribution')
axes[1,1].set_xlabel('Days since last purchase')
axes[1,1].set_ylabel('Number of customers')

plt.tight_layout()
plt.savefig('/Users/poornavenkat/Documents/GitHub/SQL-Projects/02-Ecommerce-Customer-Analytics/data/Output/rfm_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print("Visualizations saved!")
```

```python


# Save the RFM dataframe to CSV
rfm.to_csv('/Users/poornavenkat/Documents/GitHub/SQL-Projects/02-Ecommerce-Customer-Analytics/data/Output/rfm_analysis.csv', index=True)  # Keep index since customer_id is index

# Also save a version with reset index for easier use
rfm_with_id = rfm.reset_index()
rfm_with_id.to_csv('/Users/poornavenkat/Documents/GitHub/SQL-Projects/02-Ecommerce-Customer-Analytics/data/Output/rfm_analysis_with_id.csv', index=False)

print("RFM Analysis saved to ./outputs/rfm_analysis.csv")
print(f"Total customers analyzed: {len(rfm)}")
print("\nFiles saved:")
print("1. ./outputs/rfm_analysis.csv (with customer_id as index)")
print("2. ./outputs/rfm_analysis_with_id.csv (with customer_id as column)")


```

```python
# ===== GENERATE MISSING TABLEAU FILES =====

# Reload master dataset
master_df = pd.read_csv('/Users/poornavenkat/Documents/GitHub/SQL-Projects/02-Ecommerce-Customer-Analytics/data/Output/master_dataset.csv')
master_df['order_purchase_timestamp'] = pd.to_datetime(master_df['order_purchase_timestamp'])

output_path = '/Users/poornavenkat/Documents/GitHub/SQL-Projects/02-Ecommerce-Customer-Analytics/data/Output/'

# 1. Customer Geography (for map)
customer_geo = master_df[master_df['order_status'] == 'delivered'].groupby(
    ['customer_state', 'customer_city']
).agg({
    'order_id': 'count',
    'payment_total': 'sum',
    'customer_id': 'nunique'
}).reset_index()
customer_geo.columns = ['State', 'City', 'Total_Orders', 'Total_Revenue', 'Unique_Customers']
customer_geo.to_csv(output_path + 'customer_geography.csv', index=False)
print(f" customer_geography.csv saved: {len(customer_geo)} rows")

# 2. Monthly Trend (for time series)
monthly_trend = master_df[master_df['order_status'] == 'delivered'].copy()
monthly_trend['order_month'] = monthly_trend['order_purchase_timestamp'].dt.to_period('M').astype(str)
monthly_summary = monthly_trend.groupby('order_month').agg({
    'order_id': 'count',
    'payment_total': 'sum',
    'customer_id': 'nunique'
}).reset_index()
monthly_summary.columns = ['Month', 'Orders', 'Revenue', 'Customers']
monthly_summary.to_csv(output_path + 'monthly_trend.csv', index=False)
print(f" monthly_trend.csv saved: {len(monthly_summary)} rows")

# 3. Segment Summary (for KPIs)
rfm = pd.read_csv(output_path + 'rfm_analysis_with_id.csv')
segment_summary = rfm.groupby('Segment').agg({
    'Recency': 'mean',
    'Frequency': 'mean', 
    'Monetary': ['mean', 'sum', 'count']
}).reset_index()
segment_summary.columns = ['Segment', 'Avg_Recency', 'Avg_Frequency', 'Avg_Monetary', 'Total_Revenue', 'Customer_Count']
segment_summary.to_csv(output_path + 'segment_summary.csv', index=False)
print(f" segment_summary.csv saved: {len(segment_summary)} rows")

print("\n All Tableau files ready in:", output_path)
```
